```
def translate(sentence_ja):
  import MeCab

  with open('input.txt', mode='w') as file:
    m = MeCab.Tagger('-Owakati -d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd')
    file.write(m.parse(sentence_ja))

  !fairseq-interactive --path 'drive/My Drive/Colab Notebooks/100_knocks/chapter_10/results/97_result_checkpoints_-4/checkpoint5.pt' --beam 10 'drive/My Drive/Colab Notebooks/100_knocks/chapter_10/results/91_result_data-bin' < input.txt | grep '^H' | cut -f 3 > output.txt

  with open('output.txt') as file:
    sentence_en = file.readlines()[0].rstrip('\n')

  return sentence_en
```

```
import os

html = '''
<html>
<body>
<p>日本語を入力してください:</p>
<form action = "/output" method = "POST">
<p><textarea name = "input" cols = "30" rows = "5"></textarea></p>
<p><input type = "submit" value = "翻訳"></p>
</form>
</body>
</html>
'''

if not os.path.isdir('templates'):
  os.makedirs('templates')
with open('templates/input.html', mode='w') as f:
  f.write(html)
```

```
import os

html = '''
<html>
<body>
<p>翻訳結果が出ました:</p>
{{sentence_en}}
</body>
</html>
'''

if not os.path.isdir('templates'):
  os.makedirs('templates')
with open('templates/output.html', mode='w') as f:
  f.write(html)
```

```
from flask import *
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)

@app.route('/')
def input():
  return render_template('input.html')

@app.route('/output', methods = ['GET', 'POST'])
def output():
  sentence_ja = request.form['input']
  sentence_en = translate(sentence_ja)
  return render_template('output.html', sentence_en = sentence_en)

if __name__ == "__main__":
  app.run()
```
