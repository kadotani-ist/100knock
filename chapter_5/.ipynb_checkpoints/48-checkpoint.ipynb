```python
class Morph:
    def __init__(self, surface, base, pos, pos1):
        self.surface = surface
        self.base = base
        self.pos = pos
        self.pos1 = pos1
        
    def __str__(self):
        return 'surface: {0}, base: {1}, pos: {2}, po1: {3}'.format(self.surface, self.base, self.pos, self.pos1)
```

```python
class Chunk:
    def __init__(self, morphs, dst):
        self.morphs = morphs
        self.dst = dst
        self.srcs = []
        self.surfaces2 = ''
        
        for morph in self.morphs:
            if morph.pos != '記号':
                self.surfaces2 += morph.surface
        
    def __str__(self):
        return 'surfaces2: {0}, dst: {1}, srcs: {2}'.format(self.surfaces2, self.dst, self.srcs)
```

```python
fname = 'neko.txt.cabocha'
Morphs_list = []
Morphs = []
Chunks_list = []
Chunks = []
Chunk_Morphs = []
EOS_count = 0
src_dst = []
result = []

with open(fname) as file:
    for line in file:
        line = line.rstrip('\n')
        if line[0] == '*':
            if len(Chunk_Morphs) > 0:
                Chunks.append(Chunk(Chunk_Morphs, phrase_info[2][:-1]))
            Chunk_Morphs = []
            phrase_info = line.split(' ')
            if EOS_count > 0:
                for s_d in src_dst:
                    if int(s_d[1]) != -1:
                        Chunks[int(s_d[1])].srcs.append(s_d[0])
                src_dst = []
                Chunks_list.append(Chunks)
                Chunks = []
                EOS_count -= 1
            while EOS_count > 0:
                Chunks_list.append([])
                EOS_count -= 1
            src_dst.append([phrase_info[1], phrase_info[2][:-1]])
            continue
        if line == 'EOS':
            Morphs_list.append(Morphs)
            Morphs = []
            EOS_count += 1
            continue
        morpheme = line.split('\t')
        surface = morpheme[0]
        base = morpheme[1].split(',')[6]
        pos = morpheme[1].split(',')[0]
        pos1 = morpheme[1].split(',')[1]
        Morphs.append(Morph(surface, base, pos, pos1))
        Chunk_Morphs.append(Morph(surface, base, pos, pos1))
        
for Chunks in Chunks_list:
    for c in Chunks:
        flag = 0
        
        for morph in c.morphs:
            if morph.pos == '名詞':
                flag = 1   
        if flag == 1:
            noun_to_root = c.surfaces2
            c_next = c
            
            while c_next.dst != '-1':
                c_next = Chunks[int(c_next.dst)]
                noun_to_root += ' -> ' + c_next.surfaces2
            result.append(noun_to_root)
        
result
```
