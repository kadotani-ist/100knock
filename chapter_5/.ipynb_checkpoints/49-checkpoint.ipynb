```python
class Morph:
    def __init__(self, surface, base, pos, pos1):
        self.surface = surface
        self.base = base
        self.pos = pos
        self.pos1 = pos1
        
    def __str__(self):
        return 'surface: {0}, base: {1}, pos: {2}, po1: {3}'.format(self.surface, self.base, self.pos, self.pos1)
```

```python
class Chunk:
    def __init__(self, morphs, dst):
        self.morphs = morphs
        self.dst = dst
        self.srcs = []
        self.surfaces2 = ''
        
        for morph in self.morphs:
            if morph.pos != '記号':
                self.surfaces2 += morph.surface
        
    def __str__(self):
        return 'surfaces2: {0}, dst: {1}, srcs: {2}'.format(self.surfaces2, self.dst, self.srcs)
    
    def make_replaced_surfaces(self, character):
        replaced_surfaces = ''
        noun_flag = 0
        
        for morph in self.morphs:
            if morph.pos == '名詞' and noun_flag == 0:
                replaced_surfaces += character
                noun_flag = 1
            elif morph.pos == '記号' or (morph.pos == '名詞' and noun_flag == 1):
                continue
            else:
                replaced_surfaces += morph.surface
                
        return replaced_surfaces
```

```python
fname = 'neko.txt.cabocha'
Morphs_list = []
Morphs = []
Chunks_list = []
Chunks = []
Chunk_Morphs = []
EOS_count = 0
src_dst = []
result = []

with open(fname) as file:
    for line in file:
        line = line.rstrip('\n')
        if line[0] == '*':
            if len(Chunk_Morphs) > 0:
                Chunks.append(Chunk(Chunk_Morphs, phrase_info[2][:-1]))
            Chunk_Morphs = []
            phrase_info = line.split(' ')
            if EOS_count > 0:
                for s_d in src_dst:
                    if int(s_d[1]) != -1:
                        Chunks[int(s_d[1])].srcs.append(s_d[0])
                src_dst = []
                Chunks_list.append(Chunks)
                Chunks = []
                EOS_count -= 1
            while EOS_count > 0:
                Chunks_list.append([])
                EOS_count -= 1
            src_dst.append([phrase_info[1], phrase_info[2][:-1]])
            continue
        if line == 'EOS':
            Morphs_list.append(Morphs)
            Morphs = []
            EOS_count += 1
            continue
        morpheme = line.split('\t')
        surface = morpheme[0]
        base = morpheme[1].split(',')[6]
        pos = morpheme[1].split(',')[0]
        pos1 = morpheme[1].split(',')[1]
        Morphs.append(Morph(surface, base, pos, pos1))
        Chunk_Morphs.append(Morph(surface, base, pos, pos1))
        
for Chunks in Chunks_list:
    for i, c in enumerate(Chunks):
        flag = 0
        
        for morph in c.morphs:
            if morph.pos == '名詞':
                flag = 1
        if flag == 1: # 名詞句Xを発見
            x = c
            x_to_root = [x.dst]
            x_next = x
            
            while x_next.dst != '-1':
                x_next = Chunks[int(x_next.dst)]
                x_to_root.append(x_next.dst)
            for j, c2 in enumerate(Chunks[i+1:], i+1):
                flag2 = 0
                
                for morph2 in c2.morphs:
                    if morph2.pos == '名詞':
                        flag2 = 1
                if flag2 == 1: # 名詞句Yを発見
                    y = c2
                    y_to_root = [y.dst]
                    y_next = y
                    flag3 = 0
                    
                    while y_next.dst != '-1':
                        y_next = Chunks[int(y_next.dst)]
                        y_to_root.append(y_next.dst)
                    for node in x_to_root:
                        if node == str(j): # 文節iから構文木の根に至る経路上に文節jが存在する場合
                            x_to_y = x.make_replaced_surfaces('X')
                            c_next = x
                            
                            while True:
                                c_next = Chunks[int(c_next.dst)]
                                if c_next.surfaces2 == y.surfaces2:
                                    x_to_y += ' -> ' + c_next.make_replaced_surfaces('Y')
                                    break
                                else:
                                    x_to_y += ' -> ' + c_next.surfaces2
                            result.append(x_to_y)
                            flag3 = 1
                            break
                    if flag3 == 0: # 文節iと文節jから構文木の根に至る経路上で共通の文節kで交わる場合
                        for node in x_to_root:
                            if node in y_to_root:
                                x_to_node = x.make_replaced_surfaces('X')
                                y_to_node = y.make_replaced_surfaces('Y')
                                c_next = x
                                c_next2 = y
                                
                                while c_next.dst != node:
                                    c_next = Chunks[int(c_next.dst)]
                                    x_to_node += ' -> ' + c_next.surfaces2
                                while c_next2.dst != node:
                                    c_next2 = Chunks[int(c_next2.dst)]
                                    y_to_node += ' -> ' + c_next2.surfaces2
                                result.append('{0} | {1} | {2}'.format(x_to_node, y_to_node, Chunks[int(node)].surfaces2))
                                break
                        
result
```
