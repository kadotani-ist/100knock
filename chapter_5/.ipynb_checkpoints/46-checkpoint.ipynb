```python
class Morph:
    def __init__(self, surface, base, pos, pos1):
        self.surface = surface
        self.base = base
        self.pos = pos
        self.pos1 = pos1
        
    def __str__(self):
        return 'surface: {0}, base: {1}, pos: {2}, po1: {3}'.format(self.surface, self.base, self.pos, self.pos1)
```

```python
class Chunk:
    def __init__(self, morphs, dst):
        self.morphs = morphs
        self.dst = dst
        self.srcs = []
        self.surfaces2 = ''
        
        for morph in self.morphs:
            if morph.pos != '記号':
                self.surfaces2 += morph.surface
        
    def __str__(self):
        return 'surfaces2: {0}, dst: {1}, srcs: {2}'.format(self.surfaces2, self.dst, self.srcs)
```

```python
fname = 'neko.txt.cabocha'
Morphs_list = []
Morphs = []
Chunks_list = []
Chunks = []
Chunk_Morphs = []
EOS_count = 0
src_dst = []
result_list = []

with open(fname) as file:
    for line in file:
        line = line.rstrip('\n')
        if line[0] == '*':
            if len(Chunk_Morphs) > 0:
                Chunks.append(Chunk(Chunk_Morphs, phrase_info[2][:-1]))
            Chunk_Morphs = []
            phrase_info = line.split(' ')
            if EOS_count > 0:
                for s_d in src_dst:
                    if int(s_d[1]) != -1:
                        Chunks[int(s_d[1])].srcs.append(s_d[0])
                src_dst = []
                Chunks_list.append(Chunks)
                Chunks = []
                EOS_count -= 1
            while EOS_count > 0:
                Chunks_list.append([])
                EOS_count -= 1
            src_dst.append([phrase_info[1], phrase_info[2][:-1]])
            continue
        if line == 'EOS':
            Morphs_list.append(Morphs)
            Morphs = []
            EOS_count += 1
            continue
        morpheme = line.split('\t')
        surface = morpheme[0]
        base = morpheme[1].split(',')[6]
        pos = morpheme[1].split(',')[0]
        pos1 = morpheme[1].split(',')[1]
        Morphs.append(Morph(surface, base, pos, pos1))
        Chunk_Morphs.append(Morph(surface, base, pos, pos1))
        
for Chunks in Chunks_list:
    case_dic = {} # key = 動詞, value = [格, 項]のリスト(格の辞書順)
    
    for c in Chunks:
        src = c
        dst = Chunks[int(c.dst)]
        flag, flag2 = 0, 0
        
        for morph in src.morphs:
            if morph.pos == '助詞':
                particle = morph.base
                flag = 1
        for morph in dst.morphs:
            if morph.pos == '動詞':
                verb = morph.base
                flag2 = 1
                break
        if c.dst != '-1' and flag == 1 and flag2 == 1:
            if verb in case_dic.keys():
                case_dic[verb].append([particle, src.surfaces2])
                case_dic[verb].sort()
            else:
                case_dic[verb] = [[particle, src.surfaces2]]
                
    for verb, case_list in case_dic.items():
        particles = ' '.join([case[0] for case in case_list])
        terms = ' '.join([case[1] for case in case_list])
        result_list.append('{0}\t{1}\t{2}'.format(verb, particles, terms))
            
with open('results/46_result.txt', 'w') as file:
    for result in result_list:
        file.write('{0}\n'.format(result))
```
