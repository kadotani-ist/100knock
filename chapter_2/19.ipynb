```python
from itertools import groupby 

fname = 'popular-names.txt'
col1 = []

with open(fname) as file:
    for line in file:
        element = line.split()
        col1.append(element[0])
        col1.sort()
        
result = [[len(list(g)), k] for k, g in groupby(col1)]
result.sort(key=lambda r:r[0], reverse=True)

for r in result:
    print('{0} {1}'.format(r[0], r[1]))
```
