```python
import gensim

model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
import csv
import numpy as np
from sklearn.cluster import KMeans

countries = []
vectors = []
k = 5

with open('contries.csv') as file:
    reader = csv.reader(file)
    for row in reader:
        countries.append(row[0])
        vectors.append(np.squeeze(model[row]))
vectors_queue = np.array(vectors)
clusters = KMeans(n_clusters = k).fit_predict(vectors_queue)
result = [[cluster, country] for cluster, country in zip(clusters, countries)]
result.sort(key = lambda r: r[0])

for r in result:
    print('{0} {1}'.format(r[0], r[1]))
```
